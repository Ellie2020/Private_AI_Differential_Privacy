```python
entries=5000
db=torch.rand(entries) >0.5
 
def parallel_db(db, i):                #parallel_db function with db and item with i-index (to be removed) as arguments 
    return torch.cat([db[:i], db[i+1:]])
```

### 3. Creating a function for list of parallel databases. Each database is smaller of 1 entry than the original database is originated from.

```python
def list_parallel_dbs(db):
    list_dbs=[]
    for i in range(len(db)):
        list_dbs.append(parallel_db(db, i))
    return list_dbs   
```

```python
list_parallel_dbs(db)  
```
